```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
dataset = pd.DataFrame(data=data['data'], columns=data['feature_names'])
dataset
```

```python
from sklearn.model_selection import train_test_split
X = dataset.copy()
y = data['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(ccp_alpha=0.01)
clf = clf.fit(X_train, y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report

# Customize the decision tree according to the problem
dtc_model = DecisionTreeClassifier(max_depth=2)

# Decision tree function
def dt_clf(dtc_model, X_train, X_test, y_train, y_test):
    
    # Train the decision tree model
    model = dtc_model
    model.fit(X_train, y_train)
    
    # Use the trained model for predictions
    predictions = model.predict(X_test)
    predict_prob = model.predict_proba(X_test)
    
    # Display parameters in dataframe
    model_params = model.get_params()    
    df_params = pd.DataFrame(model_params.items(),columns=['', 'parameters'])
    df_params = df_params.set_index('')

    return predictions, predict_prob, df_params

    # dtc_pred, dtc_pred_prob, dtc_cr = dt_clf(dtc_model, X_train, X_test, y_train, y_test)
    
dtc_pred, dtc_pred_prob, dtc_cr = dt_clf(dtc_model, X_train, X_test, y_train, y_test)
```

```python
from sklearn import metrics 

def clf_model_metrics(y_test, y_pred):
    
    # Place metrics in a variable
    accuracy = metrics.accuracy_score(y_test, y_pred)
    precision = metrics.precision_score(y_test, y_pred)

    # Display result of the metrics
    print(f'Model accuracy is {accuracy:.2%}.')
    print(f'Model precision is {precision:.2%}.')

    return

clf_model_metrics(y_test, predictions)
```
