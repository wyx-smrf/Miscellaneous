# Function Conversion

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
```

```python
# Specify the dataset from scikit-learn
data = load_breast_cancer()

# Convert sklearn dataset to pandas dataframe
def sklearn_df_transformation(sklearn_dataset, usage='regression'):
    
    # Specified uses, raises error if input is not in the list
    usages = ['classification', 'regression']
    
    if usage not in usages:
        raise ValueError("Invalid operation type. Expected one of: %s" % usages)
    
    # Specifies the data and the column
    sklearn_df = pd.DataFrame(data=sklearn_dataset['data'], 
                      columns=sklearn_dataset['feature_names'])
    
    # Adds target to the dataframe
    sklearn_df['target'] = pd.Series(sklearn_dataset['target'])
    
    # If usage is classification, adds the class names of the targets
    if (usage == 'classification'):
        sklearn_df['class'] = pd.Categorical.from_codes(sklearn_dataset.target,
                                            sklearn_dataset.target_names)
    else:
        pass
    
    return sklearn_df

# Use the function
df = sklearn_df_transformation(data, usage='classification')
```

```python
# Specify the dataset from scikit-learn
data = load_breast_cancer()

# Return independent and dependent variables
def xy_variables(sklearn_dataset):
    features = pd.DataFrame(data=sklearn_dataset['data'],
                      columns=sklearn_dataset['feature_names'])
    target = pd.Series(data = sklearn_dataset['target'])
    
    return features, target

# Use the function
X, y = xy_variables(data)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report

# Specify the model here
dtc_model = DecisionTreeClassifier(ccp_alpha=0.01)

# Decision Tree Classifier
def dt_clf(X_train, X_test, y_train, y_test):
    
    # Train the decision tree model
    model = dtc_model
    trained_dtc = model.fit(X_train, y_train)
    
    # Use the trained model for predictions
    predictions = model.predict(X_test)
    
    # View decision probabilities of the model
    pred_probability = model.predict_proba(X_test) # Unused
    
    # View model parameters
    parameters = model.get_params()

    return (trained_dtc, predictions, parameters)

# Use the function
trained_dtc, dtc_pred, dtc_params = dt_clf(X_train, X_test, y_train, y_test)
```

```python
from sklearn import metrics 

# Return accuracy and precision of a classifier model
def clf_model_metrics(y_test, y_pred):
    
    # Place metrics in a variable
    accuracy = metrics.accuracy_score(y_test, y_pred)
    precision = metrics.precision_score(y_test, y_pred)

    # Display result of the metrics
    print(f'Model accuracy is {accuracy:.2%}.')
    print(f'Model precision is {precision:.2%}.')

# Use the function as 
clf_model_metrics(y_test, dtc_pred)
```

```python
from sklearn.metrics import confusion_matrix
import pandas as pd
import numpy as np

# Confusion matrix in dataframe format
def confusion_matrix_dataframe(y_true, y_pred):
    
    # Confusion matrix sorted ascending
    cm = confusion_matrix(y_true, y_pred) # Use for proportion result
    
    # Transform confusion matrix result to dataframe
    df = pd.DataFrame(cm, index=['malignant', 'benign'], columns=['malignant', 'benign'])
    
    return df

# Use function as
conf_mat = confusion_matrix_dataframe(y_test, dtc_pred)
conf_mat
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, dtc_pred, target_names=['malignant','benign'])) 
```

```python
def most_important_features(fitted_model_dtc, Xfeatures_df):
    
    # Column names of the fatures
    feature_names = Xfeatures_df.columns
    
    # Weight of the features in the trained model
    feature_weight = fitted_model_dtc.feature_importances_
    
    # Covert weight list of columns names to a dataframe
    feature_importance = pd.DataFrame(feature_weight, 
                                      index = feature_names).sort_values(0, ascending=False)

    # Optional: Return non-zero feautre values as list
    features_list = list(feature_importance[feature_importance[0]>0].index)
 
    # Feature plot
    feature_plot = feature_importance.head(10).plot(kind='barh')
    
    return (feature_importance.head(5), features_list)

# Use function as
key_features, feature_value_plot = most_important_features(trained_dtc, X)
```
